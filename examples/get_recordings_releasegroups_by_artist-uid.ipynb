Example to query recordings, release_groups and work entities credited to an artist entity

```python
# importing the functions
from mb_queries import *
```

```python
# get artist uids from the database
artists = get_all_artists(limit=20) #in this 20 rows from the srtist table
# A list of musicbrainz artist uids
artists
```

```python
# artist_uid for the band Audioslave (https://musicbrainz.org/artist/020bfbb4-05c3-4c86-b372-17825c262094)
artist_uid = '020bfbb4-05c3-4c86-b372-17825c262094'

# function get uids, titles and table ids for all the recordings credited to the band Audioslave
rec_uid, rec_names, rec_ids = get_all_recordings_by_artist(artist_uid)
```

```python
# Let's view this output as a table by importing it to a pandas dataframe.
recs = {"rec.uid":rec_uid,"rec.name":rec_names,"rec.id":rec_ids}
rec_df = pd.DataFrame(recs) #No need to import pandas package since we already imported it with mb_queries.py
rec_df.head(20) #first 20 rows
```

```python
# Let's query release_group uids and titles for all the release_group credited to the band Audioslave
rgroup_uid, rgroup_names = get_all_releasegroups_by_artist(artist_uid)
```

```python
rgroups = {"release_group.uid":rgroup_uid,"release_group.name":rgroup_names}
rgroup_df = pd.DataFrame(rgroups)
rgroup_df.head(10) #first 10 rows
```
